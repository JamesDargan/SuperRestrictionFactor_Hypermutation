```python

#iterate through samples 
for key in sample_directs:
    #dirlist = sample_directs[key]
    print("\n*********************")
    print ("\n parsing {0} files:".format(key))
    r1files = sample_directs[key]
    r2files = [w.replace('_R1_001.fastq.gz', '_R2_001.fastq.gz') for w in r1files]
    #r1files =[]
    #r2files = []
    #for direct in dirlist:
        
        #print("looking in {0}".format(direct))
        #dirr1files = glob.glob('{0}*R1*.fastq.gz'.format(direct))
        #for f in dirr1files:
        #    r1files.append(f)
        #dirr2files = glob.glob('{0}*R2*.fastq.gz'.format(direct))
        #for f2 in dirr2files:
        #    r2files.append(f2)
        
    assert len(r1files) == len(r2files)
    r1files.sort()
    r2files.sort()
    catf = '{0}_R1.fastq.gz'.format(key)
    if os.path.isfile(catf):
        print("\nConcatenated FASTQ file {0} already exists.".format(catf))
        print("\nFor this script, {0} would have been made by concatenating:\n\t{1}".format(
                catf, '\n\t'.join(r1files)))
    else:
        print("\n!!!!! \nCreating {0} by concatenating:\n\t{1}".format(
                catf, '\n\t'.join(r1files)))
        subprocess.check_call('zcat {0} | gzip > {1}'.format(
                ' '.join(r1files), catf), shell=True)
#now, do the same for R2 files
    catf2 = '{0}_R2.fastq.gz'.format(key)
    if os.path.isfile(catf2):
        print("\nConcatenated FASTQ file {0} already exists.".format(catf2))
        print("\nFor this script, {0} would have been made by concatenating:\n\t{1}".format(
                catf2, '\n\t'.join(r2files)))
    else:
        print("\nCreating {0} by concatenating:\n\t{1}".format(
                catf2, '\n\t'.join(r2files)))
        subprocess.check_call('zcat {0} | gzip > {1}'.format(
                ' '.join(r2files), catf2), shell=True)
```
